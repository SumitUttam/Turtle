```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
class complex:
    def __init__ (self, x, y):
        self.x = x
        self.y = y

    def __str__ (self):
        return str(self.x) + " + " + str(self.y) + "i"

    def __mul__ (self, other):
        x = self.x * other.x - self.y*other.y
        y = self.y * other.x + self.x*other.y
        return complex(x,y)

    def magnitude (self):
        return (self.x**2 + self.y**2)**(1/2)

    def __add__ (self, other):
        if type(other) in [float , int]:
            return complex(self.x+other, self.y)
        return complex(self.x+other.x, self.y+other.y )
```

```python
C = complex(0.6,0.8)
print(C)

Cs = [C*C]
x = [0.6]
y = [0.8]

for i in range(51):
    Cs.append(Cs[i]*Cs[i])
    x.append(Cs[i+1].x)
    y.append(Cs[i+1].y)

# print(x,y)

plt.plot(x,y,color='r')
plt.show()


```

```python
y = np.linspace(-2,1,3000)
x = np.linspace(-1.5,1.5,3000)

z = []

for j in x:
    temp = []
    for i in y:
        c = complex(i,j)
        m = c.magnitude()
        k = 0
        l = complex(0,0)
        while (l.magnitude() < 2 and k<100):
            k+=1
            l = c + l*l

        temp.append(k)
    z.append(temp)

# print(z)
```

```python
plt.figure(figsize=(128, 128))
plt.imshow(z,extent=[-1,1,-2,1], interpolation='bilinear')
plt.axis('scaled')
# plt.colorbar()
plt.show()
```
